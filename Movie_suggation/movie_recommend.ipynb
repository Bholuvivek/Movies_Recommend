```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv("/content/tmdb_5000_movies.csv")
credit = pd.read_csv("/content/tmdb_5000_credits.csv")
```

```python
movies.head()
```

```python
credit.head()
```

```python
movies = movies.merge(credit, on='title')
```

```python
movies.head(1)
```

```python
movies.info()
```

```python
movies=movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
movies.head(2)
```

```python
movies.shape
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast
def conversion(obj):
  a = []
  for i in ast.literal_eval(obj):
    a.append(i['name'])
  return a
```

```python
movies['genres'] = movies['genres'].apply(conversion)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(conversion)
```

```python
movies.head()
```

```python
def conversion_cast(obj):
  b = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter != 3:
      b.append(i['name'])
      counter += 1
    else:
      break
  return b
```

```python
movies['cast'] = movies['cast'].apply(conversion_cast)
```

```python
movies.head()
```

```python
def fetch_crew(obj):
  c = []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      c.append(i['name'])
      break
  return c
```

```python
movies['crew'] = movies['crew'].apply(fetch_crew)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
movies_df = movies[['movie_id', 'title', 'tags']]
```

```python
movies_df.head()
```

```python
movies_df['tags'] = movies_df['tags'].apply(lambda x:" ".join(x))
```

```python
movies_df.head()
```

```python
movies_df['tags'][0]
```

```python
movies_df['tags'] = movies_df['tags'].apply(lambda x:x.lower())
```

```python
movies_df['tags'][0]
```

```python
movies_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  a= []

  for i in text.split():
    a.append(ps.stem(i))

  return " ".join(a)
```

```python
movies_df['tags'] = movies_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vector = cv.fit_transform(movies_df['tags']).toarray()
```

```python
vector
```

```python
vector[0]
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similar = cosine_similarity(vector)
```

```python
similar
```

```python
def recommend(movie):
  movie_index = movies_df[movies_df['title'] == movie].index[0]
  distance = similar[movie_index]
  movie_list = sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]

  for i in movie_list:
    print(movies_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
